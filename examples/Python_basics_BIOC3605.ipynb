### Reference (of address) VS assignment (of value) in Python

```python
# reference of memory address
a = [1, 2, 3]
b = a
a.append(6)
print(b)
```

```python
# assignment of value
a = [1, 2, 3]
b = a.copy()
a.append(6)
print(b)
```

## Flow Control

```python
# Basic if, elif, and else
x = 7
if x < 0:
    print("It's negative")
elif x == 0:
    print("Equal to zero")
elif 0 < x < 5:
    print("Positive but smaller than 5")
else:
    print("Positive and >= to 5")
```

```python
# For loop, with break, continue
sequence = [1, 2, 0, 9, 5, 4, 6]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value

print(total_until_5)
```

```python
# List, Set, and Dictionary Comprehensions
result = []
_collection = [1, 5, 9]
_condition = lambda x: x <= 6

for value in _collection:
    if _condition(value):
        result.append(value**2)

print(result)

result2 = [value**2 for value in _collection if _condition(value)]
print(result2)
```

## Functions and modules

```python
# Basic function
def my_function(x, y):
  return x + y

result = my_function(1, 2)
result
```

```python
# Function use parameter as local variable
def func1():
    a = []
    for i in range(3):
        a.append(i)
    return a

print(func1())
print(func1())
```

```python
# Function updates out-side variable!!!
a = []
def func2():
    for i in range(3):
        a.append(i)
    return a
print(func2())
print(func2())
```

```python
# Anonymous (Lambda) Functions
equiv_anon = lambda x: x * 2
equiv_anon(5)
```
